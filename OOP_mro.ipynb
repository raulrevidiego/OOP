```python
class a:
  def metodo(self):
    print("habla a")
```

```python
class b(a):
  def metodo(self):
    print("habla b")
class c(a):
  def metodo(self):
    print("habla c")
class d(b):
  def metodo(self):
    print("habla d")
```

```python
d = d()
d.metodo()
```

El MRO es como python prioriza las herencias, lo hace por ramas, una vez seleciona una rama la prueba por completo

```python
class e (b,c):
  def metodo(self):
    print("habla e")
print(e.mro())
```

Ese seria el orden de prioridad

```python
#para llamar a un metodo especifico uso la clase y luego llamo al objeto
E = e()
b.metodo(E)
```
