El encapsulamiento es una forma de proteger cualquier elemento de una clase (variable, método, atributo,...)

```python
class miclase:
  def __init__(self):
    self._atributo_privado = "Soy un atributo privado" #En python para hacer un atributo privado se usa _
    self.__atributo_muy_privado ="Soy un atributo muy privado" #En python para hacer un atributo muy privado se usa __
```

```python
objeto = miclase()
print(objeto._atributo_privado)
```

```python
print(objeto.__atributo_muy_privado)
```

```python
#Tambien existen los set, iguales pero definen
class persona:
  def __init__(self,nombre,edad):
    self._nombre = nombre
    self._edad = edad
  def get_nombre(self):
    return self._nombre
  def set_nombre(self,nuevo_nombre):
    self._nombre = nuevo_nombre

Raul_2 = persona("Raul",23) #Creamos el objeto
nombre= Raul_2.get_nombre() #Obtenemos el nombre
print(nombre)
Raul_2.set_nombre("Raul_2") #Cambiamos el nombre
nombre = Raul_2.get_nombre()
print(nombre)
```
