```python
class persona:
  def __init__(self,nombre,edad,nacionalidad):
    self.nombre=nombre
    self.edad=edad
    self.nacionalidad=nacionalidad
  def hablar(self):
    print("Hola esto es herencia entre clases")
```

```python
class empleado(persona):
  def __init__(self,nombre,edad,nacionalidad,trabajo,salario):
    super().__init__(nombre,edad,nacionalidad)
    self.trabajo = trabajo
    self.salario = salario
roberto = empleado("Roberto",43,"aleman","camarero","1000")
roberto.hablar()
```

```python
class estudiante(persona):
  def __init__(self,nombre,edad,nacionalidad,carrera,curso):
    super().__init__(nombre,edad,nacionalidad)
    self.carrera=carrera
    self.curso=curso
  def estudiar(self):
    print("Estoy estudiando")
martin = estudiante("Martín",22,"portugués","geología","2º")
martin.estudiar()
```
