```python
#Polimorfismo
class gato():
  def sonido(self):
    return "miau"
class perro():
  def sonido(self):
    return "guau"
```

```python
gato=gato()
perro=perro()
print(gato.sonido())
print(perro.sonido())
```

```python
def hacer_sonido(animal):
  print(animal.sonido())
hacer_sonido(gato)
hacer_sonido(perro)
```

IMPORTANTE, el polimorfismo en lenguajes estaticos (jaca, c++,...) necesita herencia para funcionar
